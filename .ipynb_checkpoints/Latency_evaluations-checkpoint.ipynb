### Methodology

- Run this node on this commit:

- In simulation, run simulation for ~30 seconds through forest, first with rviz on and with rviz off

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns
import pandas as pd
```

```python
latency_columns = list(["msg", "seq", "stamp", "latency (ms)", "rviz_on"])
latency_df = pd.DataFrame(columns=latency_columns)
with open('/home/peteflo/latency/sim_latency_8_sim_beforerviz.txt') as f:
    lines = f.readlines()
    for line in lines:
        words = line.split()
        words[1] = int(words[1])
        words[2] = float(words[2])
        words[3] = 1000*float(words[3])
        latency_df_add = pd.DataFrame([words], columns=latency_columns)
        latency_df = latency_df.append(latency_df_add, ignore_index=True)
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(20,6))
ax = sns.violinplot(x="msg", y="latency (ms)", data=latency_df)
```

```python
plt.figure(figsize=(20,6))
ax = sns.violinplot(x="msg", y="latency (ms)", data=latency_df, order=["/twist"])
```
