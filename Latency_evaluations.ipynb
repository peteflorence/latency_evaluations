## Methodology

- Measured latencies coming into planning node (motion_selector_node) via the following method:
    - Inside of callback for each message, coming into the node, measure the difference of ros::Time::now() and the timestamp of the message

- Can see this implemented on this branch that was used: https://bitbucket.org/fla-draper-mit/motion_primitives/branch/pf-measure-latency

- In simulation, ran simulator for ~20 seconds through forest, first with rviz on and with rviz off

- From bagged data, ran replay script for duration of 10 m/s demo flight through warehouse (actual measured top speed was 9.79 m/s).  (flight_2017_02_20_15_53)

- *Note that each message was measured both with rviz off, and with rviz on (blue and green respectively)

- For the simulator, a 10-core i7 was used
- For the bagged data, a quad-core i7 was used

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns
import pandas as pd
```

```python
latency_columns = list(["msg", "seq", "stamp", "latency (ms)", "rviz_on"])
latency_df = pd.DataFrame(columns=latency_columns)
with open('/home/peteflo/latency/sim_latency_8_sim_beforerviz.txt') as f:
    lines = f.readlines()
    for line in lines:
        words = line.split()
        words[1] = int(words[1])
        words[2] = float(words[2])
        words[3] = 1000*float(words[3])
        latency_df_add = pd.DataFrame([words], columns=latency_columns)
        latency_df = latency_df.append(latency_df_add, ignore_index=True)
```

```python
with open('/home/peteflo/latency/sim_latency_9_sim_afterrviz.txt') as f:
    lines = f.readlines()
    for line in lines:
        words = line.split()
        words[1] = int(words[1])
        words[2] = float(words[2])
        words[3] = 1000*float(words[3])
        latency_df_add = pd.DataFrame([words], columns=latency_columns)
        latency_df = latency_df.append(latency_df_add, ignore_index=True)
```

## Latencies in simulation

In our Unity/Drake simulator, following latencies are measured for all messages coming in:

```python
sns.set_style("whitegrid")
plt.figure(figsize=(20,6))
ax = sns.violinplot(x="msg", y="latency (ms)", hue="rviz_on", data=latency_df, split=True)
```

Zooming in on the three low-latency messages:

```python
plt.figure(figsize=(20,6))
ax = sns.violinplot(x="msg"
                    , y="latency (ms)", 
                    hue="rviz_on", data=latency_df, split=True, order=["/twist", "/pose", "/local_goal"])
```
